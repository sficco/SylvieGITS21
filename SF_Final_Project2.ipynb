```python
import os
import pandas as pd
import geopandas as gp
import numpy as np
import matplotlib.pyplot as plt
```

```python
midterm = pd.read_csv('alldata.csv')
```

```python
midterm
```

```python
midterm['lon'] = [-122.33, -122.60, -118.23, 104.99, 118.89, 112.07, 87.62, 90.19, 96.79, 71.05, 77.03, 80.19, 124.20, 118.08, 112.04, 114.62, 95.67, 94.16, 70.30, 71.41]
midterm['lat'] = [47.61, 37.77, 34.05, 39.73, 40.76, 33.44, 41.87, 38.62, 32.77, 42.36, 38.90, 25.76, 41.75, 45.32, 43.49, 32.69, 39.04, 45.55, 41.70, 41.82]
```

```python
midterm
```

```python
cities = gp.GeoDataFrame(midterm, geometry = gp.points_from_xy(midterm.lon, midterm.lat))
cities
```

```python
filepath = 'C:\\Users\\sficc\\gitstuff\\SylvieGITS21\\finalproject\\states.shp'
states = gp.read_file(filepath)
```

```python
states.head()
```

```python
states = states.to_crs("epsg:4326") #Mercator-projection
```

```python
cities
```

```python
fig, ax = plt.subplots(figsize=(20,20))
states.plot(ax=ax, color="lightgray", edgecolor="grey", linewidth=0.4, legend = True)
cities.plot(ax=ax,color='red', markersize='Poor_Days', alpha=1.0, categorical=False, legend = True)
ax.axis()
plt.title('Air Quality Index Poor Days: 2020', size=20)
plt.savefig('finalproject/Aqi2020.png')
#plt.axis()
#plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,20))
states.plot(ax=ax, color="lightgray", edgecolor="grey", linewidth=0.4, legend = True)
cities.plot(column = 'Zone', ax=ax, markersize='Poor_Days10', alpha=1.0, categorical=False, legend = True)
ax.axis()
plt.title('Air Quality Index Poor Days: 2010', size=20)
plt.savefig('finalproject/Aqi2010.png')
#plt.axis()
#plt.show()
```

```python
import folium

# Make an empty map
m = folium.Map(location=[34,-100], tiles="OpenStreetMap", zoom_start=4)

# Show the map
m
```

```python
import folium

for i in range(0,len(cities)):
   folium.Circle(
      location=[cities.iloc[i]['lat'], cities.iloc[i]['lon']],
      popup=cities.iloc[i]['City'],
      radius=float(cities.iloc[i]['Poor_Days'])*2000,
      color='red',
      fill=True,
      fill_color='red'
   ).add_to(m)

# Show the map again
m
```

```python
m.save('finalproject/aqidata2020.html')
```
